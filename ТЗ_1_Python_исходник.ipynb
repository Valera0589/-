Вам предстоит поработать со списком возрастов посетителей фитнес-клубов age_lst

**Задача 1.** Рассчитайте средний возраст по всему списку age_lst

**Задача 2.** Рассчитайте средний возраст для посетителей, которые моложе 50 лет

**Задача 3.** Оберните предыдущий код в функцию, которая на вход принимает список и пороговое значение 

**Задача 4.** Рассчитайте среднее значение для всех элементов списка, за исключением тех, которые лежат вне интервала [Q1-1,5*IQR; Q3+1,5*IQR]

**Задача 5 (со звёздочкой).** Оберните предыдущий код в функцию, которая принимает на вход список и рассчитывает средний возраст, исключая из расчётов выбросы

```python
age_lst = [23, 45, 18, 80, 22, 25, 34, 37, 41, 88, 19, 20, 18, 45, 41, 39, 31, 32, 22, 28, 37 ]
```

```python
#Задача 1
avg_age = sum(age_lst) / len(age_lst)
print(f"Средний возвраст посетителя {round(avg_age , 2)} лет")
```

```python
print(sorted(age_lst))
```

```python
#Задача 2
sum_age = 0
cnt_age = 0
for i in age_lst :
    if i < 50 :
        sum_age += i
        cnt_age += 1
avg_age2 = sum_age / cnt_age
print(f"Средний возвраст посетителя до 50 лет {round(avg_age2 , 2)}")
```

```python
#Задача 3
def calc_avg(lst , thresh) :
    sum_age = 0
    cnt_age = 0
    for i in lst :
        try :
            if i < thresh :
                sum_age += i
                cnt_age += 1
        except TypeError:
            print(f'{i} - нечисловое значение')
    try :
        avg_age = sum_age / cnt_age
        print(f'Средний возвраст посетителя до {thresh} лет')
        return round(avg_age , 2)
    except :
        print(f'Список должен содержать хотя бы одно значение меньше {thresh}')
```

```python
calc_avg(age_lst , 50)
```

```python
q1 = np.percentile(age_lst , 25)
q2 = np.percentile(age_lst , 50)
q3 = np.percentile(age_lst , 75)
p95 = np.percentile(age_lst , 95)
print(q1 , q2 , q3 , p95)
```

```python
iqr = q3-q1
lower_bound = q1-1.5*iqr
upper_bound = q3+1.5*iqr
print(lower_bound)
print(upper_bound)
```

```python
#Задача 4
sum_age = 0
cnt_age = 0
for i in age_lst :
    if (i < upper_bound) and (i > lower_bound) :
        sum_age += i
        cnt_age += 1
avg_age2 = sum_age / cnt_age
print(f"Средний возвраст посетителя {round(avg_age2 , 2)}")
```

```python
#Задача 5
def calc_avg_adv(lst) :
    q1 = np.percentile(lst , 25)
    q3 = np.percentile(lst , 75)
    iqr = q3-q1
    lower_bound = q1-1.5*iqr
    upper_bound = q3+1.5*iqr
    sum_age = 0
    cnt_age = 0
    for i in lst :
        if (i < upper_bound) and (i > lower_bound) :
            sum_age += i
            cnt_age += 1
    avg_age = sum_age / cnt_age
    return (round (avg_age , 2))
```

```python
calc_avg_adv(age_lst)
```
