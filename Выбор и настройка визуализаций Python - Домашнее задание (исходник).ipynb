```python
##############################################   Задание 2.1   #################################################################

##### Импортните csv-файл под названием Purchase_List
##### Приведите date_purchase к формату даты с помощью функции to_datetime
##### Создайте колонку датафрейма с месяцесем оплат (числом, от 1 до 12)
##### Сделайте группировку: количество оплат и сумма оплат для каждого месяца (создайте новый группировочный датафрейм)
##### Постройте визуализацию: для каждого месяца покупки по левой оси постройте сумму оплат, а по правой - количество оплат
##### Не забудьте оформление графика (заглавие, подпись осей, раздельная легенда, сетка)

import pandas as pd
import datetime as dt

purchase = pd.read_csv('Purchase_List.csv' )
purchase['date_purchase'] = pd.to_datetime(purchase ['date_purchase'])
purchase['month_come'] = purchase['date_purchase'].dt.month

# Сгруппированный DataFrame
purchase_month = purchase.groupby(['month_come']).agg(
    amt_pay = ('amt_pay' , 'sum'),
    cnt_pay = ('amt_pay' , 'count')
).reset_index()

# Строим график
Fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(purchase_month['month_come'] , purchase_month['cnt_pay'] , 'r' , label = 'Сумма оплат')
ax.set_ylabel('Кол-во оплат')
ax.set_xlabel('Месяц')
ax.set_title('Лист оплат')
ax.grid()
ax.legend(loc = 'upper right')
ax1 = ax.twinx()
ax1.bar(purchase_month['month_come'] , purchase_month['amt_pay'], alpha = 0.5, color = 'b', label = 'Кол-во оплат')
ax1.legend(loc = 'upper left')
ax1.ticklabel_format(style = 'plain' ,  axis = 'y')
ax1.set_ylabel('Сумма оплат')

plt.show()
```
