### Задание 1

Создайте функцию func_calc, которая будет на вход брать словарь и суммировать его значения.

```python
 def func_calc(dct):
    dsum = 0
    for value in dct.values():
        dsum += value
    return dsum
```

Протестируйте вашу функцию на словаре dct_1

```python
dct_1 = {'a':1, 'b':6, 'c':7, 'd':8, 'e':2}
```

```python
print(func_calc(dct_1))
```

### Задание 2

С помощью конструкции try-except дополните вашу функцию, чтобы отработать потенциальные не-цифровые значения словаря. Обратите внимание, что сумма должна считаться даже если в значениях есть не числовые элементы, и не числовые элементы должны обрабатываться принтом.

```python
def func_calc(dct):
    dsum = 0
    for value in dct.values():
        try :
            dsum += value
        except TypeError :
            print ("Ошибка типа данных")
    return dsum
```

Протестируйте вашу функцию на словаре dct_2

```python
dct_2 = {'a':1, 'b':[1,2,3], 'c':7, 'd':8, 'e':'f'}
```

```python
print(func_calc(dct_2))
```

### Задание 3

Добавьте в функцию расчет среднего значения словаря (разделите сумму на количество значений). Данная цифра также должна быть в return функции.

Добавьте проверку на наполненность словаря с помощью конструкции try-except: функция должна принтить "Словарь должен быть не пустой!" в случае если ей был передан пустой словарь в качестве аргумента.

```python
def func_calc(dct):
    dsum = 0
    count = 0
    if not dct :
        return ("Словарь не должен быть пустой")
    for value in dct.values():
        try :
            dsum += value
            count += 1
        except TypeError :
            print ("Ошибка типа данных")
    avg_dct = dsum / count
    return(dsum , avg_dct)
```

Протестируйте вашу функцию на словаре dct_2

```python
dct_2 = {'a':1, 'b':[1,2,3], 'c':7, 'd':8, 'e':'f'}
```

```python
print(func_calc(dct_2))
```

### Задание 4

Добавьте в функцию еще один аргумент: пороговое значение thr (от англ. Threshold).
    
В расчете суммы и среднего должны участвовать только те значения словаря, которые превышают (строго) thr. Задайте параметру thr значение по умолчанию: 0.

Обратите внимание на последовательность вложенности: for -> try -> if

```python
def func_calc(dct, thr = 0):
    dsum = 0
    count = 0
    if not dct :
        return ("Словарь не должен быть пустой")
    for value in dct.values():
        try :
            if value < thr :
                dsum += value
                if count < thr :
                    count += 1
        except TypeError :
            print ("Ошибка типа данных")
    if count == 0 :
        return ("Нет значений , превышающих порог")
    avg_dct = dsum / count
    return(dsum , avg_dct)
```

Протестируйте вашу функцию на словаре dct_4

```python
dct_4 = {'a':1, 'b':[1,2,3], 'c':7, 'd':-8, 'e':'f'}
```

```python
print(func_calc(dct_4))
```
